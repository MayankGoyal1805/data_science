```python
import numpy as np
import matplotlib as mp
import matplotlib.image as mpimg

img = mpimg.imread("Doraemon-1.jpg")
print(type(img))
print(img.shape)
# with open("temp.txt","w") as file :
#     file.write(np.array2string(img,threshold=np.inf,separator=","))
cop = img.copy()
cop[:,:,:2] = 0 
mpimg.imsave("blue.jpg",cop)
```

```python
import matplotlib.pyplot as plt

plt.imshow(img)
```

```python
avg = img.mean(axis=2)
with open("temp.txt","w") as file:
    file.write(np.array2string(avg,separator=",",threshold=np.inf))
plt.imshow(avg,cmap="gray")
avg.shape
```

```python
avg = img.mean(axis=2)

avg_rgb = np.zeros_like(img)
avg_rgb[:,:,0] = avg
avg_rgb[:,:,1] = avg
avg_rgb[:,:,2] = avg

plt.imshow(avg_rgb)
```

```python
# Calculate true luminance
luminance = 0.299 * img[:,:,0] + 0.587 * img[:,:,1] + 0.114 * img[:,:,2]

# Create RGB version where all channels are equal
grayscale_rgb = np.zeros_like(img)
grayscale_rgb[:,:,0] = luminance  # Red channel
grayscale_rgb[:,:,1] = luminance  # Green channel
grayscale_rgb[:,:,2] = luminance  # Blue channel

plt.imshow(grayscale_rgb)
```
