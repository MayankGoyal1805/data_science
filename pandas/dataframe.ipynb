```python
import numpy as np
import pandas as pd
```

```python
d = {
    "one": pd.Series([1.0,2.0,3.6],index=["a","b","c"]),
    "two": pd.Series([1.0,2.0,3.0,4.0],index=["a","b","c","d"])
}

df = pd.DataFrame(d)
df
```

```python
pd.DataFrame(d,index=["d","b","a"])
```

```python
pd.DataFrame(d,index=["d","b","a"],columns=["two","three"])
```
