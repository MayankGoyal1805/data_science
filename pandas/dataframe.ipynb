```python
import numpy as np
import pandas as pd
```

```python
a = [pd.Series([1.0,2.0,3.6],index=["a","b","c"],name="huha"),pd.Series([1.0,2.0,3.0,4.0],index=["a","b","c","d"],name="try")]
pd.DataFrame(a)
```

```python
d = {
    "one": pd.Series([1.0,2.0,3.6],index=["a","b","c"]),
    "two": pd.Series([1.0,2.0,3.0,4.0],index=["a","b","c","d"])
}

df = pd.DataFrame(d)
df
```

```python
pd.DataFrame(d,index=["d","b","a"])
```

```python
pd.DataFrame(d,index=["d","b","a"],columns=["two","three"])
```

```python
print(df.index)
```

```python
print(df.columns)
```

```python
d1 = {
    "one": [1.0,2.0,3.0,4.0],
    "two": [4.0,3.0,2.0,1.0]
}

pd.DataFrame(d1,index=["a","b","c","d"])
```

DataFrames from structured or record arrays 

```python
data = np.zeros((2,),dtype=[("A","i4"),("B","f4"),])
```

```python
data
```

```python
data = np.zeros((2,), dtype=[("A", "i4"), ("B", "f4"), ("C", "S10")])
data
```

```python
pd.DataFrame(data)
```

```python
data[:] = [(1, 2.0, "Hello"), (2, 3.0, "World")]
```

```python
pd.DataFrame(data)
```

```python
pd.DataFrame(data, index=["first", "second"])
```

```python
pd.DataFrame(data,columns=["C","B","A"])
```

From a list of dicts

```python
data2 = [{"a":1,"b":2},{"a":5,"b":10,"c":20}]
pd.DataFrame(data2)
```

```python
df = pd.DataFrame(data2,index=["first","second"],columns=["a","b","c"])
df
```

```python
df.loc["first"]
```

From a dict of tuples

```python
pd.DataFrame(
    {
        ("a","b"):{("A","B"):1, ("A","C"):2},
        ("a","a"):{("A","C"):3, ("A","B"):4},
        ("a","c"):{("A","B"):5, ("A","C"):6},
        ("b","a"):{("A","C"):7, ("A","B"):8},
        ("b","b"):{("A","D"):9, ("A","B"):10}
    }
)
```
