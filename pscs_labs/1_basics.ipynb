## 2.1. Compute the Mean (Average)

Given a list of numbers: `[12, 15, 20, 22, 24]`, write a Python program to compute the mean using the formula:

$$Mean = \frac{1}{n}\sum_{i=1}^{n}x_{i}$$

**Instructions:**
* Do not use `sum()` or `mean()`.
* Use a `for` loop to compute the sum manually.

```python
# List of numbers
data = [12, 15, 20, 22, 24]

# Initialize sum variable
total_sum = 0
n = len(data)

# Compute sum manually using a for loop
for x in data:
    total_sum += x

# Compute mean
mean_val = total_sum / n

print(f"Computed Mean: {mean_val}")
```

## 2.2. Compute the Variance and Standard Deviation

Using the same data from Q2.1, write a Python program to compute the variance and standard deviation using the formulas:

**Variance:**
$$Variance (\sigma^2) = \frac{1}{n} \sum_{i=1}^{n} (x_i - \mu)^2$$

**Standard Deviation:**
$$Standard~Deviation (\sigma) = \sqrt{Variance}$$

**Instructions:**
* First calculate the mean manually (or reuse from 2.1).
* Use the mean to compute the variance and standard deviation.

```python
# Reuse mean from previous step
mu = mean_val
n = len(data)

squared_diff_sum = 0

# Calculate sum of squared differences manually
for x in data:
    squared_diff_sum += (x - mu) ** 2

# Compute Variance
variance = squared_diff_sum / n

# Compute Standard Deviation
std_dev = variance ** 0.5

print(f"Variance: {variance}")
print(f"Standard Deviation: {std_dev}")
```

## 2.3. Estimate Empirical Probability

Given a list of binary outcomes (e.g., success/failure): `[1, 0, 1, 1, 0, 0, 1, 1]`, write a program to estimate the probability of success (1) using:

$$P(success) = \frac{Number~of~successes}{Total~trials}$$

```python
outcomes = [1, 0, 1, 1, 0, 0, 1, 1]

success_count = 0
total_trials = len(outcomes)

# Count successes
for outcome in outcomes:
    if outcome == 1:
        success_count += 1

# Calculate probability
p_success = success_count / total_trials

print(f"Probability of Success: {p_success}")
```

## 2.4. Grade Classification and Probability Estimation

You are given a list of student scores out of 100: `[85, 72, 90, 66, 58, 45, 38, 93, 77, 50]`.

Classify each score into a letter grade based on the following rules:
* **A:** 80 and above
* **B:** 60 to 79
* **C:** 40 to 59
* **F:** below 40

**Instructions:**
1.  Use `if-elif-else` statements to classify each score into its grade.
2.  Count how many students fall into each grade.
3.  Compute the probability of getting each grade, using:
    $$P(Grade) = \frac{Count~of~Grade}{Total~Scores}$$
4.  Print the probabilities in the format: $P(A)=0.2$, etc.

```python
scores = [85, 72, 90, 66, 58, 45, 38, 93, 77, 50]

# Initialize counts
counts = {'A': 0, 'B': 0, 'C': 0, 'F': 0}
total_scores = len(scores)

# Classify and count
for s in scores:
    if s >= 80:
        counts['A'] += 1
    elif s >= 60:
        counts['B'] += 1
    elif s >= 40:
        counts['C'] += 1
    else:
        counts['F'] += 1

# Compute and print probabilities
for grade, count in counts.items():
    prob = count / total_scores
    print(f"P({grade}) = {prob}")
```

## 2.6. Identify Invalid or High Values

You are given a list of time measurements in seconds: `[10, 12, 15, 1, 0, 20, 18, 25, 5, 30]`.

Write a program to:
* Count how many values are **not valid** (zero or negative).
* Count how many values are **valid and greater than 15**.
* Use the `not` and `and` operators appropriately.

```python
measurements = [10, 12, 15, 1, 0, 20, 18, 25, 5, 30]

invalid_count = 0
high_valid_count = 0

for m in measurements:
    # Check if not valid (assuming valid implies > 0 based on context of "zero or negative")
    if not (m > 0):
        invalid_count += 1
    
    # Check if valid AND greater than 15
    if (m > 0) and (m > 15):
        high_valid_count += 1

print(f"Invalid values: {invalid_count}")
print(f"Valid values > 15: {high_valid_count}")
```

## 2.8. Estimate Conditional Probability

Given a list of session data represented as pairs `[('user1', 1), ('user2', 0), ('user3', 1), ...]`, where 1 indicates a purchase and 0 indicates no purchase.

Assume only users who visited the site between 6pm and 9pm are labeled (i.e., the list provided contains only those users). Estimate:

$$P(purchase | visited~between~6pm~and~9pm) = \frac{No~of~purchases}{No~of~total~sessions}$$

**Instructions:** Count purchases manually using a loop. Use only string labels and integers.

## 2.9. Empirical Distribution Function (PMF)

Given a list of integers representing the number of customer support tickets per day for 15 days: `[2, 3, 2, 4, 3, 3, 1, 2, 4, 4, 2, 2, 1, 3, 4]`.

Compute the empirical PMF:
$$P(X=k) = \frac{Number~of~times~k~appears}{Total~observations}, ~~ \text{for } k \in \{1, 2, 3, 4\}$$

**Instructions:** Use a loop to count how many times each value appears.

```python
tickets = [2, 3, 2, 4, 3, 3, 1, 2, 4, 4, 2, 2, 1, 3, 4]
total_obs = len(tickets)

# Identify unique values k
unique_k = sorted(list(set(tickets)))

print("Empirical PMF:")
for k in unique_k:
    count_k = 0
    for t in tickets:
        if t == k:
            count_k += 1
    
    prob_k = count_k / total_obs
    print(f"P(X={k}) = {prob_k:.4f}")
```

## 2.10. Empirical Cumulative Distribution Function (CDF) Using Lists

You are given a list containing the ages of visitors to a website (in years):
`[19, 22, 25, 30, 21, 24, 20, 18, 28, 27]`

**Instructions:**
1.  Sort the list in ascending order using the `sort()` function.
2.  For each value $x$ in the sorted list, compute the empirical cumulative distribution function (CDF) using:
    $$P(X \le x) = \frac{Number~of~values \le x}{Total~number~of~values}$$
3.  Store the results in a list of tuples of the form `(x, P(X <= x))`.
4.  Print all values in a tabular format showing the age and the cumulative probability.

```python
ages = [19, 22, 25, 30, 21, 24, 20, 18, 28, 27]
ages.sort()

n_values = len(ages)
cdf_list = []

for i, x in enumerate(ages):
    # Since list is sorted, number of values <= x is simply the current index (i) + 1
    count_le_x = i + 1
    prob_le_x = count_le_x / n_values
    cdf_list.append((x, prob_le_x))

print(f"{'Age':<10} {'CDF P(X<=x)':<15}")
print("-" * 25)
for x, prob in cdf_list:
    print(f"{x:<10} {prob:<15.2f}")
```

## 2.11. Function to Run a Monte Carlo Experiment

You want to estimate the value of $\pi$ using random sampling. Write a Python function `def estimate_pi(trials):` to:

1.  Randomly generate points in the unit square: $x, y \in [0, 1]$.
2.  Count how many of those points fall inside the quarter unit circle, i.e., satisfy:
    $$x^2 + y^2 \le 1$$
3.  Classify each point as **'Inside'** if $x^2 + y^2 \le 1$ and **'Outside'** otherwise.
4.  Use the following formula to return the estimated value of $\pi$:
    $$\hat{\pi} = 4 \times \frac{points~inside~circle}{total~trials}$$
5.  Plot the sampling points using the `matplotlib` library with points inside the circle in **blue** and points outside in **red**. Include a legend and title "Monte Carlo Simulation of $\pi$".

```python
import random
import matplotlib.pyplot as plt

def estimate_pi(trials):
    inside_count = 0
    
    # Lists for plotting
    x_inside, y_inside = [], []
    x_outside, y_outside = [], []
    
    for _ in range(trials):
        # 1. Randomly generate points
        x = random.random()
        y = random.random()
        
        # 2. Check condition
        if x**2 + y**2 <= 1:
            inside_count += 1
            x_inside.append(x)
            y_inside.append(y)
        else:
            x_outside.append(x)
            y_outside.append(y)
            
    # 4. Estimate Pi
    pi_estimate = 4 * (inside_count / trials)
    
    # 5. Plotting
    plt.figure(figsize=(6, 6))
    plt.scatter(x_inside, y_inside, color='blue', s=1, label='Inside')
    plt.scatter(x_outside, y_outside, color='red', s=1, label='Outside')
    plt.title(f"Monte Carlo Simulation of $\pi$ (Estimate: {pi_estimate})")
    plt.legend()
    plt.xlim(0, 1)
    plt.ylim(0, 1)
    plt.show()
    
    return pi_estimate

# Run the function
estimate_pi(1000)
```
