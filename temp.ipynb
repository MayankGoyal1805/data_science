```python
class DocFormat(Enum):
    PDF = 1
    TXT = 2
    MD = 3
    HTML = 4

def convert_format(content,from_format,to_format):
    match (from_format,to_format):
        case (from_format.MD,to_format.HTML):
            if content[0] == "#":
                content = content[1:].strip()
                return f"<h1>{content}</h1>"
```
